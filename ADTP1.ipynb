## ETL (Extraccion, Transformacion, Carga)


### 1.Extracción (Extract)

```python
from google.colab import drive
import pandas as pd
#Montar drive
drive.mount('/content/drive', force_remount=True)

ruta_archivo = '/content/drive/MyDrive/Analisis de Datos Informatorio/dengue2018.csv'

try:
  df = pd.read_csv(ruta_archivo,)
  print('Extracción completada')
except FileNotFoundError:
  print('El archivo no se encuentra en la ruta especificada')
```

### 3. Carga(Load)

## EDA (Análisis exploratorio de datos)


```python
# listado de casos grupo_edad_desc entre 10 a 14 años con provincia_nombre y evento_nombre
filtered_df = df[df['grupo_edad_desc'] == 'de 10 a 14 anos']
display(filtered_df[['provincia_nombre', 'evento_nombre', 'cantidad_casos']])
```

```python
# Provincias con más casos de enfermedad
provincia_evento_casos = df.groupby(['provincia_nombre', 'evento_nombre'])['cantidad_casos'].sum().reset_index()

# Ordenar los resultados por el total de casos en orden descendente
sorted_provincia_evento_casos = provincia_evento_casos.sort_values(by='cantidad_casos', ascending=False)

# Mostrar la lista ordenada
display(sorted_provincia_evento_casos)
```

```python
#Filtrado condicional (booleano)
#condicion: provincia CABA , enfermedad Dengue
condicion= (df['provincia_nombre'] =='CABA') & (df['evento_nombre'] == 'Dengue')
print(condicion)

```
