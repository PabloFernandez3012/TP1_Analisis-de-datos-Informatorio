## ETL (Extraccion, Transformacion, Carga)


### 1.Extracción (Extract)

```python
from google.colab import drive
import pandas as pd
#Montar drive
drive.mount('/content/drive', force_remount=True)

ruta_archivo = '/content/drive/MyDrive/Analisis de Datos Informatorio/dengue2018.csv'
with open (ruta_archivo, 'rb') as archivo:
  print(archivo.read())

try:
  df = pd.read_csv(ruta_archivo,)
  print('Extracción completada')
except FileNotFoundError:
  print('El archivo no se encuentra en la ruta especificada')
```

### 2.Transformacion (Transform)

```python
#Muestra las primeras 10 filas del dataframe
df.head(10)
```

```python
#informacion del dataframe
df.info()
```

```python
#generar estadísticas descriptivas de las columnas numéricas
df.describe()
```

```python
#renombrar columna ano por año
df.rename(columns={'ano':'año'}, inplace=True)
print(df['año'])
```

```python
#Lista de enfermedades de estudio
print(df['evento_nombre'].unique())

#imprimir lista de provincia, enfermedad y cantidad de casos
print(df.groupby('provincia_nombre')['evento_nombre'].value_counts())
```

### 3. Carga(Load)

```python
#definir la ruta de salida para guardar el nuevo csv
ruta_salida = '/content/drive/MyDrive/Analisis de Datos Informatorio/dengue2.csv'

#guardar el dataframe en un nuevo archivo csv en Drive
df.to_csv(ruta_salida, index=False)

print('Carga completada')
```
